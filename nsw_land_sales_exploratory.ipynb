```python
import pandas as pd
```

```python
NSW_LAND_VALUE_FILE_PATH = "./data/NSW_Land_Sales_Data"
CSV_FORMAT = "*.csv"

LV_2019 = "data/NSW_Land_Sales_Data/LV_20190801/088_LAND_VALUE_DATA_20190801.csv"
LV_2024 = "data/NSW_Land_Sales_Data/LV_20240301/088_LAND_VALUE_DATA_20240301.csv"
LV2019_df = pd.read_csv(LV_2019)
LV2024_df = pd.read_csv(LV_2024)
```

```python
LV2019_df.info()
print(LV2019_df.nunique())
```

```python
def concatAddress(row: pd.Series):
    return f'{row["UNIT NUMBER"] + "/" if not pd.isnull(row["UNIT NUMBER"]) else ""} {row["HOUSE NUMBER"] if not pd.isnull(row["HOUSE NUMBER"]) else ""} {row["STREET NAME"]}, {row["SUBURB NAME"]}'

def mergeAllLandValue(row: pd.Series):
    maxRecord = 5
    result = {}
    for i in range(1,maxRecord + 1):
        landValueColumn = f'LAND VALUE {i}'
        if landValueColumn in row:
            result[row[f'BASE DATE {i}'].split('/')[2]] = row[f'LAND VALUE {i}']
    return result
```

```python
address_df = LV2019_df.apply(concatAddress, axis=1)
landValue_df = LV2019_df.apply(mergeAllLandValue, axis=1)
LV2019_df["ADDRESS"] = address_df
LV2019_df["LAND VALUE"] = landValue_df
filtered_2019_df = LV2019_df[["PROPERTY ID", "PROPERTY DESCRIPTION", "ADDRESS", "POSTCODE", "DISTRICT NAME", "DISTRICT CODE", "LAND VALUE"]]
```

```python
filtered_2019_df
```

```python
address_df = LV2024_df.apply(concatAddress, axis=1)
landValue_df = LV2024_df.apply(mergeAllLandValue, axis=1)
LV2024_df["ADDRESS"] = address_df
LV2024_df["LAND VALUE"] = landValue_df
filtered_2024_df = LV2024_df[["PROPERTY ID", "PROPERTY DESCRIPTION", "ADDRESS", "POSTCODE", "DISTRICT NAME", "DISTRICT CODE", "LAND VALUE"]]
```

```python
filtered_2024_df
```

```python
merged_df = pd.merge(filtered_2024_df, filtered_2019_df, on="PROPERTY ID", how="outer")
```

```python
def processMergedRows(row: pd.Series):
    columns = ["PROPERTY DESCRIPTION", "ADDRESS", "POSTCODE", "DISTRICT NAME", "DISTRICT CODE", "LAND VALUE"]
    for column in columns:
        if column == "LAND VALUE":
            if not pd.isnull(row["LAND VALUE_y"]) and not pd.isnull(row["LAND VALUE_x"]):
                row["LAND VALUE_x"].update(row["LAND VALUE_y"])        
            elif not pd.isnull(row["LAND VALUE_y"]) and pd.isnull(row["LAND VALUE_x"]):
                row["LAND VALUE_x"] = row["LAND VALUE_y"]
        else:
            if pd.isnull(row[f'{column}_x']) and not pd.isnull(row[f'{column}_y']):
                row[f'{column}_x'] = row[f'{column}_y']
    return row
```

```python
mergedLand_df = merged_df.apply(processMergedRows, axis=1)

mergedLand_df = mergedLand_df.rename(columns={"PROPERTY DESCRIPTION_x": "PROPERTY DESCRIPTION", "ADDRESS_x": "ADDRESS", "POSTCODE_x": "POSTCODE",
       "DISTRICT NAME_x": "DISTRICT NAME", "DISTRICT CODE_x": "DISTRICT CODE"})

mergedLand_df = mergedLand_df[["PROPERTY ID", "PROPERTY DESCRIPTION", "ADDRESS", "POSTCODE", "DISTRICT NAME", "DISTRICT CODE", "LAND VALUE"]]
mergedLand_df.head()
```
